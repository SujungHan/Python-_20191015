```python
from tensorflow import keras
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# 붓꽃 데이터 읽어 들이기
iris_data = pd.read_csv("data/iris.csv", encoding="utf-8")
```

```python
# 붓꽃 데이터를 레이블과 입력 데이터로 분리하기
y_labels = iris_data.loc[:,"class"]
x_data = iris_data.loc[:,["sepal_length","sepal_width","petal_length","petal_width"]]
```

```python
# 레이블 데이터를 One-hot 형식으로 변환하기
labels = {
    'Iris-setosa': [1, 0, 0], 
    'Iris-versicolor': [0, 1, 0], 
    'Iris-virginica': [0, 0, 1]
}
y_nums = np.array(list(map(lambda v : labels[v] , y_labels)))
x_data = np.array(x_data)
```

```python
# 학습 전용과 테스트 전용 분리하기
x_train, x_test, y_train, y_test = train_test_split(x_data, y_nums, train_size=0.8)
```

```python
type(x_train)
```

```python
# 모델의 구조 정의하기 --- (*3)
model = keras.models.Sequential([
    keras.layers.Dense(10, activation='relu', input_shape=(4,)),
    keras.layers.Dense(3, activation='softmax')
])
```

```python
model.summary()
```

```python
# 모델 구축하기
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy'])
```

```python
# 학습 실행하기
model.fit(x_train, y_train, batch_size=20, epochs=300)
```

```python
# 모델 평가하기 --- (*6)
score = model.evaluate(x_test, y_test, verbose=1)
print('정답률=', score[1], 'loss=', score[0])
```
